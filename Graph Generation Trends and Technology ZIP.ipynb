```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#NOx Emissions by Energy Technology
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_excel('/mnt/data/[EPA] Emmissions Trends CLEANED GA.xlsx')

# Reshape the data if necessary
# Assuming your Excel has columns: Year, Electric Generating Units, Off-Road & Industrial, Other, Vehicles
df_melted = df.melt(id_vars='Year', var_name='Source', value_name='NOx Emissions')

# Set up the figure
fig, ax = plt.subplots(figsize=(14, 7))

# Define custom color mapping to match the graph
colors = {
    'Electric Generating Units': '#B55200',
    'Off-Road & Industrial': '#009688',
    'Other': 'gray',
    'Vehicles': '#F4A300'
}

# Line styles (solid vs dashed)
linestyles = {
    'Electric Generating Units': '-',
    'Off-Road & Industrial': '-',
    'Other': '-',
    'Vehicles': '--'
}

# Plot each source
for source in df_melted['Source'].unique():
    subset = df_melted[df_melted['Source'] == source]
    ax.plot(
        subset['Year'],
        subset['NOx Emissions'],
        label=source,
        color=colors[source],
        linestyle=linestyles[source],
        marker='o'
    )

# Titles and labels
ax.set_title('Georgia NOₓ Emissions Trends by Source (2002-2019)', fontsize=20, weight='bold')
ax.set_xlabel('Year', fontsize=14)
ax.set_ylabel('NOₓ Emissions (millions of tons)', fontsize=14)

# Legend
ax.legend(title='Source', fontsize=12, title_fontsize=13)

# Tight layout
plt.tight_layout()

# Show the plot
plt.show()

```

```python

#Difference from Georgia Median Home Price by Energy Technology
# # Load the data
df = pd.read_csv('/mnt/data/Main ZIP Code Based Power Plant Table.csv')

# Sort by the "Difference from GA Median" column
df_sorted = df.sort_values('Difference from GA Median')

# Set up the figure
fig, ax = plt.subplots(figsize=(14, 8))

# Define a diverging colormap based on the values
colors = np.where(
    df_sorted['Difference from GA Median'] < 0,
    plt.cm.Blues(np.linspace(0.5, 1, df_sorted[df_sorted['Difference from GA Median'] < 0].shape[0])),
    plt.cm.Reds(np.linspace(0.5, 1, df_sorted[df_sorted['Difference from GA Median'] > 0].shape[0]))
)

# Merge the two color arrays back into one list in the original data order
color_list = []
neg_idx = 0
pos_idx = 0
for val in df_sorted['Difference from GA Median']:
    if val < 0:
        color_list.append(plt.cm.Blues(0.5 + 0.5 * neg_idx / max(1, df_sorted[df_sorted['Difference from GA Median'] < 0].shape[0]-1)))
        neg_idx += 1
    else:
        color_list.append(plt.cm.Reds(0.5 + 0.5 * pos_idx / max(1, df_sorted[df_sorted['Difference from GA Median'] > 0].shape[0]-1)))
        pos_idx += 1

# Plot horizontal bar chart
bars = ax.barh(
    df_sorted['Energy Technology'],
    df_sorted['Difference from GA Median'],
    color=color_list,
    edgecolor='none'
)

# Vertical dashed line at 0
ax.axvline(x=0, color='black', linestyle='--', linewidth=2)

# Set titles and labels
ax.set_title('Difference from Georgia Median Home Price by Energy Technology', fontsize=18)
ax.set_xlabel('Difference from GA Median ($)', fontsize=14)
ax.set_ylabel('Energy Technology', fontsize=14)

# Clean up spines (optional)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Layout
plt.tight_layout()
plt.show()
```
