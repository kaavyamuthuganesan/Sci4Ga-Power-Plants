# clean energy generation file

```python
import pandas as pd
```

```python
df= pd.read_csv(r"C:\Users\kaavy\OneDrive\Documents\Sci4GA Internship\GA_Generation_with_Location_By_Year.csv")
```

```python

# Drop rows where 'fuel2002' is not 'ALL' but 'primeMover' is 'ALL' to get rid of duplicate entries
df1 = df.drop(df[(df['fuel2002'] != 'ALL') & (df['primeMover'] == 'ALL')].index)

# Display the updated DataFrame
df1.head()

```

```python
# Group by 'plantName', 'Year', and 'fuelTypeDescription' and aggregate the 'generation' and 'gross-generation'
# Keep the 'Latitude' and 'Longitude' columns
aggregated_df = df1.groupby(['plantName', 'Year', 'fuelTypeDescription', 'Latitude', 'Longitude']).agg({
    'generation': 'sum',
    'gross-generation': 'sum'
}).reset_index()


# Display the aggregated DataFrame
aggregated_df.head()
```

```python
pivot_df = aggregated_df.pivot_table(
    index=['plantName', 'Year', 'Latitude', 'Longitude'],
    columns='fuelTypeDescription',
    values='gross-generation',
    aggfunc='sum'
).reset_index()

# Display the pivoted DataFrame
pivot_df.head()
```

```python
pivot_df.fillna(0, inplace=True)

# Display the updated DataFrame
pivot_df.head()
```

```python
pivot_df.columns
```

```python
# List of columns to calculate percentage for
fuel_columns = ['Distillate Fuel Oil', 'Hydroelectric Conventional', 'Hydroelectric Pumped Storage', 
                'Municiapl Landfill Gas', 'Natural Gas', 'Nuclear', 'Other', 'Petroleum Coke', 
                'Residual Fuel Oil', 'Solar', 'Waste Oil and Other Oils', 'Wood Waste Solids', 
                'other renewables']

# Calculate percentage for each column
for column in fuel_columns:
    pivot_df[column + ' Percentage'] = (pivot_df[column] / pivot_df['Total']) * 100

# Display the updated DataFrame
pivot_df.head()
```

```python
missing_values_df = pivot_df[pivot_df.isnull().any(axis=1)]
missing_values_df.head()
```

```python
unique_plant_names = missing_values_df['plantName'].unique()
print(unique_plant_names)
```

```python
pivot_df.dropna(inplace=True)

# Display the updated DataFrame
pivot_df.head()
```

```python
cleaned_energy_generation_data_year = pivot_df.copy()
```

```python
cleaned_energy_generation_data_year.to_csv('cleaned_energy_generation_data_year.csv', index=False)
```

```python
# Get unique years from the DataFrame
unique_years = cleaned_energy_generation_data_year['Year'].unique()

# Loop through each year and create a separate DataFrame
for year in unique_years:
    # Filter the DataFrame for the current year
    df_year = cleaned_energy_generation_data_year[cleaned_energy_generation_data_year['Year'] == year]
    
    # Save the DataFrame to a CSV file
    df_year.to_csv(f'cleaned_energy_generation_data_{year}.csv', index=False)
```

```python
import pandas as pd
```

```python
gen14 = pd.read_csv('cleaned_energy_generation_data_2014.csv')
gen15 = pd.read_csv('cleaned_energy_generation_data_2015.csv')
gen16 = pd.read_csv('cleaned_energy_generation_data_2016.csv')
gen17 = pd.read_csv('cleaned_energy_generation_data_2017.csv')
gen18 = pd.read_csv('cleaned_energy_generation_data_2018.csv')
gen19 = pd.read_csv('cleaned_energy_generation_data_2019.csv')
gen20 = pd.read_csv('cleaned_energy_generation_data_2020.csv')
gen21 = pd.read_csv('cleaned_energy_generation_data_2021.csv')
gen22 = pd.read_csv('cleaned_energy_generation_data_2022.csv')
gen23 = pd.read_csv('cleaned_energy_generation_data_2023.csv')
```

```python
fuel_columns = ['Coal', 'Distillate Fuel Oil', 'Hydroelectric Conventional', 'Hydroelectric Pumped Storage', 
                'Municiapl Landfill Gas', 'Natural Gas', 'Nuclear', 'Other', 'Petroleum Coke', 
                'Residual Fuel Oil', 'Solar', 'Waste Oil and Other Oils', 'Wood Waste Solids', 
                'other renewables', 'Total']

# List of DataFrames
dataframes = [gen14, gen15, gen16, gen17,gen18, gen19, gen20, gen21, gen22, gen23]

# Dictionary to store total sums for each year
yearly_totals = {}

# Sum up the fuel columns for each DataFrame and store in the dictionary
for df in dataframes:
    year = df['Year'].iloc[0]
    total_fuel_sums = {column: df[column].sum() for column in fuel_columns}
    yearly_totals[year] = total_fuel_sums

# Convert the dictionary to a DataFrame
yearly_totals_df = pd.DataFrame(yearly_totals).T
# Display the DataFrame
print(yearly_totals_df)

```

```python
yearly_totals_df.to_csv('yearly_totals.csv', index=True)
```
